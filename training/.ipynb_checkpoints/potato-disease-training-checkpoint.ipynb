# Potato Disease Classification

### Import all the Dependencies
These are the necessary libraries and modules for the script. TensorFlow and Keras are used for building and training the neural network, and other libraries like NumPy and Matplotlib are used for data manipulation and visualization.

```python
import numpy as np
import os
from tensorflow.keras import models, layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import img_to_array
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras import backend as K
```

### Set all the Constants
These are constants and configurations used throughout the script, such as learning rate, batch size, image size, and directory paths.

```python
INIT_LR = 1e-3
BATCH_SIZE = 32
EPOCHS = 50
IMAGE_SIZE = 256
default_image_size = tuple((IMAGE_SIZE, IMAGE_SIZE))
image_size = 0
data_dir = "../data/"
CHANNELS=3
AUTOTUNE = tf.data.AUTOTUNE
```

## Initializing, Exploring & Partioning the Dataset
This function is designed to split a TensorFlow dataset into training, validation, and test subsets based on specified percentages.

### Function to Split Dataset

Dataset should be bifurcated into 3 subsets, namely:
1. Training: Dataset to be used while training
2. Validation: Dataset to be tested against while training
3. Test: Dataset to be tested against after we trained a model

```python
def get_dataset_partitions_tf(ds, train_split=0.8, val_split=0.1, test_split=0.1, shuffle=True, shuffle_size=10000):
    assert (train_split + test_split + val_split) == 1
    
    ds_size = ds.cardinality().numpy()
    
    if shuffle:
        ds = ds.shuffle(shuffle_size, seed=12)
    
    train_size = int(train_split * ds_size)
    val_size = int(val_split * ds_size)
    
    train_ds = ds.take(train_size)    
    val_ds = ds.skip(train_size).take(val_size)
    test_ds = ds.skip(train_size).skip(val_size)
    
    return train_ds, val_ds, test_ds
```

### Reading and Partitioning the Dataset
We create a Tensorflow Dataset Object and directly read it from the directory using `image_dataset_from_directory` and then split it using the function we created above. Reads the dataset from the specified directory, and then partitions it into training, validation, and test sets using the function defined earlier.

```python
dataset = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  seed=123,
  image_size=default_image_size,
  batch_size=BATCH_SIZE
)


train_ds, val_ds, test_ds = get_dataset_partitions_tf(dataset)
```

Prints the class names and the number of classes in the dataset.

```python
class_names = dataset.class_names
n_classes = len(class_names)
print(n_classes, class_names)
```

### Cache, Shuffle, and Prefetch the Dataset
Caches, shuffles, and prefetches the training, validation, and test datasets to optimize training performance.

```python
train_ds = train_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().shuffle(1000).prefetch(buffer_size=AUTOTUNE)
```

## Building the Model

### Creating a Layer for Resizing and Normalization
Sequential model for resizing and rescaling images.
Before we feed our images to network, we should be resizing it to the desired size. 
Moreover, to improve model performance, we should normalize the image pixel value (keeping them in range 0 and 1 by dividing by 256).
This should happen while training as well as inference. Hence we can add that as a layer in our Sequential Model.

```python
resize_and_rescale = tf.keras.Sequential([
  layers.experimental.preprocessing.Resizing(IMAGE_SIZE, IMAGE_SIZE),
  layers.experimental.preprocessing.Rescaling(1./255),
])
```

### Data Augmentation
Sequential model for data augmentation.
Data Augmentation is needed when we have less data, this boosts the accuracy of our model by augmenting the data.

```python
data_augmentation = tf.keras.Sequential([
  layers.experimental.preprocessing.RandomFlip("horizontal_and_vertical"),
  layers.experimental.preprocessing.RandomRotation(0.2),
])
```

#### Checking what is the expected dimension order for channel

```python
input_shape = (IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
batch_input_shape = (BATCH_SIZE, IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
chanDim = -1
if K.image_data_format() == "channels_first":
    input_shape = (CHANNELS, IMAGE_SIZE, IMAGE_SIZE)
    batch_input_shape = (BATCH_SIZE, CHANNELS, IMAGE_SIZE, IMAGE_SIZE)
    chanDim = 1
```

### Model Architecture
Defines a Convolutional Neural Network (CNN) model using Keras with convolutional, max-pooling, and dense layers.
We use a CNN coupled with a Softmax activation in the output layer. We also add the initial layers for resizing, normalization and Data Augmentation.

NN Explanation:

Resizing and Rescaling Layers:

resize_and_rescale,
data_augmentation,

These are sequential layers defined earlier. The resize_and_rescale layer resizes the input images to the specified size, and the data_augmentation layer performs data augmentation to increase the variety of training data.

Convolutional and MaxPooling Layers:

layers.Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape),
layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, kernel_size=(3, 3), activation='relu'),
layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, (3, 3), activation='relu'),
layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, (3, 3), activation='relu'),
layers.MaxPooling2D((2, 2)),
layers.Conv2D(64, (3, 3), activation='relu'),
layers.MaxPooling2D((2, 2)),

These are a series of convolutional and max-pooling layers. Convolutional layers (Conv2D) apply filters to the input to extract features, and max-pooling layers (MaxPooling2D) downsample the spatial dimensions of the feature maps. This pattern helps the network learn hierarchical representations.

Flatten Layer:

layers.Flatten(),

The Flatten layer flattens the 3D output to a 1D array before feeding it into the densely connected layers.

Dense (Fully Connected) Layers:

layers.Dense(64, activation='relu'),
layers.Dense(n_classes, activation='softmax'),

These are fully connected (dense) layers. The first dense layer has 64 units and uses ReLU activation. The final dense layer has units equal to the number of classes in the dataset, and it uses softmax activation to output probabilities for each class.

Building the Model:

model.build(input_shape=batch_input_shape)

This line builds the model by specifying the input shape. It is necessary to call build to initialize the weights of the layers.

This architecture represents a typical CNN structure for image classification, starting with convolutional and pooling layers to capture local features, followed by fully connected layers for higher-level representations and class predictions. The activation functions (ReLU and softmax) introduce non-linearity into the model.

```python
model = models.Sequential([
    resize_and_rescale,
    data_augmentation,
    layers.Conv2D(32, kernel_size = (3,3), activation='relu', input_shape=input_shape),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64,  kernel_size = (3,3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(n_classes, activation='softmax'),
])

model.build(input_shape=batch_input_shape)
```

### Model Summary

```python
model.summary()
```

### Compiling the Model
We use `adam` Optimizer, `SparseCategoricalCrossentropy` for losses, `accuracy` as a metric

optimizer='adam': This specifies the optimization algorithm used during training. In this case, the Adam optimizer is employed. Adam is a popular optimization algorithm that adapts the learning rates of individual parameters, making it effective for a wide range of tasks.
Loss Function:

loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False): This sets the loss function used to measure the model's performance. The chosen loss function is Sparse Categorical Crossentropy. It is commonly used for classification problems where there are two or more classes, and the classes are mutually exclusive. The from_logits=False indicates that the model's output has already undergone a softmax activation, so there's no need to apply it again within the loss calculation.
Metrics:

metrics=['accuracy']: This specifies the metric(s) used to evaluate the performance of the model. In this case, it uses accuracy, which is a common metric for classification problems. Accuracy measures the proportion of correctly classified samples.

```python
model.compile(
    optimizer='adam',
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=False),
    metrics=['accuracy']
)
```

### Training the Network

```python
history = model.fit(
    train_ds,
    batch_size=BATCH_SIZE,
    validation_data=val_ds,
    verbose=1,
    epochs=EPOCHS,
)
```

### Testing the Model

```python
print("[INFO] Calculating model accuracy")
scores = model.evaluate(test_ds)
print(f"Test Accuracy: {round(scores[1],4)*100}%")
```

### Plotting the Accuracy and Loss Curves

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
